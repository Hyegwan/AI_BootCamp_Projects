# 데이터 수 줄이기

## 데이터 로드

```python
import pandas as pd
from tqdm import tqdm
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
try:
    ori_path = 'C:\\Users\\User\\Desktop\\AIB_13\\CP1\\data\\만개의 레시피.csv'
except:
    ori_path = '/content/drive/MyDrive/CP1/만개의 레시피.csv'
ori = pd.read_csv(ori_path, encoding='cp949')
ori = ori.dropna()
ori = ori.reset_index(drop=True)
ori.shape
```

```python
try:
    craw_path = 'C:\\Users\\User\\Desktop\\AIB_13\\CP1\\data\\ingredient.parquet'
except:
    craw_path = '/content/drive/MyDrive/CP1/ingredient.parquet'
craw = pd.read_parquet(craw_path, engine='fastparquet')
craw.shape
```

불러온 두 데이터를 합쳐줍니다.

```python
df = pd.concat([ori, craw['ingredient']], axis=1)
df.shape
```

# 특정 기준으로 로우수 제거

```python
df['CKG_INBUN_NM'].unique()
```

1~2인분의 데이터만 추출합니다.

```python
df = df.loc[(df['CKG_INBUN_NM']=='1인분') | (df['CKG_INBUN_NM'] == '2인분')]
df.shape
```

각 연도만 추출합니다.

```python
df['year'] = df['FIRST_REG_DT'].apply(lambda x : int(str(x)[:4]))
df.shape
```

```python
df['INBUN_NUM'] = df['CKG_INBUN_NM'].apply(lambda x : int(x[0]))
```

```python
df.shape
```

각 연도의 분포입니다.

```python
plt.figure(figsize=(10,5))
sns.countplot(data=df, x='year')
plt.show()
```

연도와 조회수의 분포입니다.

```python
plt.figure(figsize=(5,5))
sns.scatterplot(data=df, x='year', y='SRAP_CNT')
plt.show()
```

각 연도와 추천수의 분포입니다.

```python
plt.figure(figsize=(5,5))
sns.scatterplot(data=df, x='year', y='RCMM_CNT')
plt.show()
```

조회수의 분포입니다.

```python
plt.figure(figsize=(5,5))
sns.histplot(data=df, x='INQ_CNT', bins=10000)
plt.show()
```

```python
years = df['year'].unique()
print(df.shape)
df_counts = []
for year in years:
    df_count = df.loc[df['year']==year]['RGTR_ID'].count()
    df_counts.append(df_count)
```

```python
inq_2000_df = df.loc[df['INQ_CNT']>=2000]
print(inq_2000_df.shape)
inq_counts = []
for year in years:
    inq_count = inq_2000_df.loc[inq_2000_df['year']==year]['RGTR_ID'].count()
    inq_counts.append(inq_count)
```

각 연도별 조회수 차이입니다.

```python
for i in range(len(years)):
    print(f"{years[i]} 전: {df_counts[i]} 후: {inq_counts[i]} 차이: {df_counts[i]-inq_counts[i]}")
```

```python
copy1 = df.loc[(df['year'] <= 2017) & (df['INQ_CNT'] > 2000)].copy()
copy2 = df.loc[(df['year'] > 2017)]
new_df = pd.concat([copy1, copy2])
```
